```python
# Iterate through all pages up to the maximum page number
for page in tqdm(range(1,int(max_pages)+1)):
    # Generate the URL for the current page
    link = base_link + str(page)
    # Make a GET request to the page with a 15 second timeout
    response = requests.get(link, timeout = 15, headers = user_agent, verify=False)
    # Check if the server returned a successful response
    if response.status_code == 200:
        # If successful, parse the HTML content using BeautifulSoup
        soup = BeautifulSoup(response.content, 'html.parser')
        # Find all 'p' elements with class 'review-roaster', which contain links to coffee reviews
        coffee_links = soup.find_all('p',{'class':'review-roaster'})
        # Append the URLs of each coffee review to the list
        for coffee_link in coffee_links:
            coffee_list.append(coffee_link.find('a')['href'])      
    else:
        # If unsuccessful, print an error message
        print('Server is down!')
    # Pause for 0.6 seconds to avoid overloading the server with requests
    sleep(0.6)
```
